```python
#Rectifying error in values of 'Height' column as height values in the dataset are incorrect.
import random
ht=[]
#creating a list of random numbers between 150 and 180
for i in range(0,458):
    v=random.randint(150,180)
    ht.append(v)
ht
```

**1.a)Number of employees in each team**

```python
teamwise_count=pd.DataFrame(df['Team'].value_counts()).reset_index()
teamwise_count.columns=['Team','No of employees']
teamwise_count
```

**1.b)Percentage splitting of employees in each team with respect to the total employees**

```python
#finding percentage of employees w.r.t total employees
team_percent=[]
for i in range(0,30):  
    team_percent.append(round((teamwise_count['No of employees'].iloc[i]/sum(teamwise_count['No of employees']))*100,2))
teamwise_count['Percentage']=team_percent
teamwise_count
```

**Bar chart of percentage of employees in each team**

```python
x = teamwise_count['Percentage']
y = teamwise_count['Team']

fig, ax = plt.subplots(figsize=(12, 8), dpi=100)
bar_width = 0.8 

ax.barh(y, x, edgecolor='black', height=bar_width)

ax.set_xlabel("Percentage")
ax.set_ylabel("Teams")
ax.set_title("Percentage of Employees in Each Team")


ax.set_yticks(np.arange(len(y)))
ax.set_yticklabels(y)

plt.tight_layout()
plt.show()
```

**2. Segregating the employees w.r.t different positions**

```python
# Extract unique positions from dataframe and create a seperate emp_position dataframe
unique_positions = df['Position'].unique()
emp_position = pd.DataFrame(columns=unique_positions)

# Iterate over the positions and assign employee names to corresponding columns
for position in unique_positions:
    employee_names = df.loc[df['Position'] == position, 'Name']
    if len(employee_names) > 0:
        emp_position[position] = employee_names.reset_index(drop=True)
emp_position
```

**3.Age group to which most employees belong to**

```python
#finding minimum and maximum of age
print("Minimum age: ",df['Age'].min())
print("Maximum age: ",df['Age'].max())
```

```python
#Classifying employees to different age groups and finding the count of each age group
bins = [20, 25,30,35,40]
df['Age Group'] = pd.cut(df['Age'], bins=bins)
age_group_counts = df['Age Group'].value_counts()
print("Employees of different age groups:\n",age_group_counts)

# Age group with the maximum count
max_age_group = age_group_counts.idxmax()
print(f"The age group with the maximum number of employees is {max_age_group}")

# Visualizing the age groups and their employee counts
age_group_counts.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Employee Distribution by Age Group')
plt.show()
```

**5.Correlation between age and salary** 

```python
df1=df[['Age','Salary']]
df1
```

```python
#Visual representation of age-salary correlation using heatmap

correlation = df1[['Age', 'Salary']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True,cmap='coolwarm',linewidths=0.5)
plt.title("Correlation between Age and Salary")
plt.show()
```

Above heat map shows that the correlation between Age and Salary values is 0.21. This suggests that there is a positive, but a weak correlation between the variables.Hence, cannot form a definitive conclusion about the relationship between age and salary.There may be other factors, which may influence salary and also,there can be considerable variation in salary within each age group. This can also be further verified using the following salary distribution line graph.

**Line graph showing average salary distribution across different age groups of the dataset**

```python
# Grouping the data by 'Age Group' and calculate average salary
grouped_data = df.groupby('Age Group')['Salary'].mean()
# Extracting age groups and average salaries
age_groups = grouped_data.index.astype(str)
average_salaries = grouped_data.values

# Plotting the line chart
plt.plot(age_groups, average_salaries,marker='^')
plt.xlabel('Age Group')
plt.ylabel('Average Salary in Millions')
plt.title('Average salary distribution across age groups')
plt.grid(True)
plt.show()
```
